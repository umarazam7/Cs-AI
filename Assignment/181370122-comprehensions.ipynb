# 1. Create a list using list comprehension that contains numbers from 1 to 1000

```python
new_list = [i for i in range(1 , 1000)]
print(new_list)
```

# 2. Create a list using list comprehension that contains numbers from 1 to 1000 with step size of 2.

```python
new_list = [i for i in range(1 , 1000 , 2)]
print(new_list)
```

# 3. Create a list using list comprehension that contains numbers from 1 to 1000 that are even.

```python
new_list = [i for i in range(1 , 1000) if i%2 == 0]
print(new_list)
```

# 4. Create a list using list comprehension that contains numbers from 1 t0 1000 that are odd.

```python
new_list = [i for i in range(1 , 1000) if i%2 != 0]
print(new_list)
```

# 5. Create a list using list comprehension that contains numbers from 1 to 1000 that are divisible by
8.

```python
new_list = [i for i in range(1 , 1000) if i%8 ==0]
print(new_list)
```

# 6. Create a list using list comprehension that contains all those numbers that have a 5 in them.
HINT: Use in operator by type casting number to string. E.g. “8” in str(789)

```python
new_list = [i for i in range(1 , 1000) if '5' in str(i)]
print(new_list)
```

# 7. Create a list using list comprehension that contains squares of first 200 integers.

```python
new_list = [i**2 for i in range(1 , 200)]
print(new_list)
```

# 8. Create a list using list comprehension that contains squares of first 200 even integers.

```python
new_list = [i**2 for i in range(1 , 200) if i%2==0]
print(new_list)
```

# 9. Consider the following list:
fruits = [‘apple’, ‘banana’, ‘cherry’, ‘blueberry’, ‘mango’,
‘watermelon’]
a. Create a new list using list comprehension that contains all fruits’ names in uppercase.
Use upper() method of strings to achieve this.
b. Create a new list using list comprehension that contains all fruits’ names that contains
‘a’ in the name. Transform to uppercase. HINT: Use membership operator and upper().

```python
fruits = ['apple', 'banana', 'cherry', 'blueberry', 'mango',
'watermelon']
Fruits=[x.upper() for x in fruits]
print(Fruits)
```

```python
Fruits=[x.upper()  for x in fruits if 'a' in x]
print(Fruits)
```
